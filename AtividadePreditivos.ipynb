```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_wine
from sklearn.datasets import load_boston

from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.model_selection import GridSearchCV

from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import LinearRegression
```

# Classificação

```python
baseC = load_wine()
```

```python
X_train, X_test, y_train, y_test = train_test_split(baseC['data'], baseC['target'], test_size=0.33)
```

## KNN

```python
param_grid = {
    'n_neighbors': range(1,40,2),
    'metric': ['euclidean', 'minkowski', 'chebyshev'], 
}
```

```python
grid = GridSearchCV(KNeighborsClassifier(),param_grid, verbose = 3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

## Naive

```python
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
print("Acurácia:", metrics.accuracy_score(y_test, pred))
```

## SVM

```python
param_grid = {
    'C': [0.1,10,100], 
    'gamma': [0.1,0.01,0.001], 
    'kernel': ['rbf', 'poly', 'sigmoid', 'linear']
} 
```

```python
grid = GridSearchCV(SVC(),param_grid, verbose = 3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

# MLP

```python
param_grid = {
    'hidden_layer_sizes': [(13,13), (8,8), (8,)],
    'activation': ['logistic', 'tanh', 'relu'],
    'solver': ['lbfgs', 'adam'],
    'learning_rate_init': [0.001, 0.01, 0.1],
    'learning_rate': ['constant','adaptive'],
    'max_iter': [500, 1000, 5000],
    'momentum': [0.5, 0.8, 0.9], 
}
```

```python
grid = GridSearchCV(MLPClassifier(),param_grid, verbose = 3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

# Regressão

```python
baseR = load_boston()
```

```python
X_train, X_test, y_train, y_test = train_test_split(baseR['data'], baseR['target'], test_size=0.33)
```

## KNN

```python
param_grid = {
    'n_neighbors': range(1,40,2),
    'metric': ['euclidean', 'minkowski', 'chebyshev'], 
}
```

```python
grid = GridSearchCV(KNeighborsRegressor(),param_grid, verbose = 3, scoring='neg_mean_absolute_error')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
knn = KNeighborsRegressor(metric = 'euclidean', n_neighbors= 3)
```

```python
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
```

```python
print("erro: ", metrics.mean_absolute_error(y_test, pred))
```

## SVM

```python
param_grid = {
    'C': [0.1,10,100], 
    'kernel': ['rbf','linear']
} 

grid = GridSearchCV(SVR(),param_grid, verbose = 3, scoring='neg_mean_absolute_error')

grid.fit(X_train,y_train)
```

```python
print(grid.best_params_)
```

```python
svm = SVR(C = 10, kernel = 'linear')
```

```python
svm.fit(X_train,y_train)
```

```python
pred = svm.predict(X_test)
print("erro: ", metrics.mean_absolute_error(y_test, pred))
```

## MLP

```python
param_grid = {
    'hidden_layer_sizes': [(13,13), (13,)],
    'activation': ['logistic', 'tanh'],
    'solver': ['lbfgs', 'adam'],
    'learning_rate_init': [0.001, 0.01, 0.1],
    'max_iter': [500, 1000],
}
grid = GridSearchCV(MLPRegressor(),param_grid, verbose = 3, scoring='neg_mean_absolute_error')

grid.fit(X_train,y_train)
```

```python
print(grid.best_params_)
```

```python
mlp = MLPRegressor(activation= 'logistic', hidden_layer_sizes= (13, 13), learning_rate_init= 0.01, max_iter= 1000, solver= 'lbfgs')
```

```python
mlp.fit(X_train,y_train)
```

```python
pred = mlp.predict(X_test)
print("erro: ", metrics.mean_absolute_error(y_test, pred))
```

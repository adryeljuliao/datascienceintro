```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer() #Essa base de dados é apresentada sob forma de dicionário
```

```python
print(cancer['DESCR'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer['data'], cancer['target'], test_size=0.33)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
model = MLPClassifier(hidden_layer_sizes=(16,16), activation = 'logistic', solver='lbfgs', learning_rate_init = 0.3, max_iter =500, momentum = 0.8, verbose = True)
```

```python
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn import metrics
```

```python
acuracia = metrics.accuracy_score(predictions, y_test)
```

```python
print(acuracia)
```

```python
param_grid = {
    'hidden_layer_sizes': [(16,16), (16,2), (30,)],
    'activation': ['logistic', 'tanh', 'relu'],
    'solver': ['lbfgs', 'adam'],
    'learning_rate_init': [0.001, 0.01, 0.1],
    'learning_rate': ['constant','adaptive'],
    'max_iter': [500, 1000, 5000],
    'momentum': [0.5, 0.8, 0.9], 
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(MLPClassifier(),param_grid, verbose = 3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
acuracia = metrics.accuracy_score(grid_predictions, y_test)
```

```python
print(acuracia)
```

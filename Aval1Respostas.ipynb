Essas bases apresentam os dados referentes as turmas de Lógica Computacional e Algoritmos e Programação Orientada a Objetos dos referidos anos.


```python
import pandas as pd

lca2014 = pd.read_csv('LCA2014.csv')

lca2015 = pd.read_csv('LCA2015.csv')

lca2016 = pd.read_csv('LCA2016.csv')

lca2017 = pd.read_csv('LCA2017.csv')

lca2018 = pd.read_csv('LCA2018.csv')
```

```python
lca = pd.concat([lca2014, lca2015, lca2016, lca2017, lca2018], keys=['2014', '2015', '2016', '2017', '2018'], names=['Turma'])
```

```python
porTurma = lca.groupby('Turma')
medias = porTurma.mean()
medias.drop('MatriculaFake', axis = 1, inplace = True)
```

**3-Apresente o percentual de aprovação (APR e APRN) de cada turma de LCA:**

*Dicas: Você pode criar uma função para isso e usar o map para gerar a Serie abaixo*

```python
def percentualAprovacao(x):
    return (resultadoPorTurma[x]['APRN'] + resultadoPorTurma[x]['APR'])/resultadoPorTurma[x].sum()
```

```python
percAprov = pd.Series(map(percentualAprovacao,['2014','2015', '2016', '2017', '2018']), index=['2014','2015', '2016', '2017', '2018'])
```

```python
percAprov
```

**Outra forma de fazer a questão 3**

```python
lca[(lca['Resultado'] == 'APRN') | (lca['Resultado'] == 'APR')].groupby('Turma')['Resultado'].count()/ lca.groupby('Turma')['Resultado'].count()
```

```python
def percentualAbandono(x):
    return resultadoPorTurma[x]['REMF']/resultadoPorTurma[x].sum()
```
